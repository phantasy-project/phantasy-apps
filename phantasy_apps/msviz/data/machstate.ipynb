```python
df_traj_x = df.xs('BPM-X', level=1)
df_traj_y = df.xs('BPM-Y', level=1)
df_pha = df.xs('BPM-PHA', level=1)
df_mag = df.xs('BPM-MAG', level=1)

df_traj_x.reset_index(inplace=True)
df_traj_x.set_index('D####', inplace=True)
df_traj_y.reset_index(inplace=True)
df_traj_y.set_index('D####', inplace=True)
df_pha.reset_index(inplace=True)
df_pha.set_index('D####', inplace=True)
df_mag.reset_index(inplace=True)
df_mag.set_index('D####', inplace=True)
```
