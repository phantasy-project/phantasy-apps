```python
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
from phantasy_apps.settings_manager import fetch_mach_state
```

```python
nshot = 5
df = fetch_mach_state(rate=1, nshot=nshot)
```

```python
df
```

```python
from datetime import datetime
import time

t0 = time.time()
datetime.fromtimestamp(t0).strftime("%Y-%m-%dT%H:%M:%S.%f")
datetime.strptime("2021-05-06T10:47:32.097", "%Y-%m-%dT%H:%M:%S.%f")
```

```python
# Add a new column for D####
index0 = df.index.get_level_values(0)
dnum = index0.str.extract(r'.*(D[0-9]{4}).*')
df['D####'] = dnum[0].to_numpy()
```

```python
df
```

```python
df.info()
```

```python
df.to_csv("/tmp/test1.csv")
```

```python
df_traj_x = df.xs('traj-x', level=1)
df_traj_y = df.xs('traj-y', level=1)
df_phase = df.xs('phase', level=1)
df_energy = df.xs('energy', level=1)

df_traj_x.reset_index(inplace=True)
df_traj_x.set_index('D####', inplace=True)
df_traj_y.reset_index(inplace=True)
df_traj_y.set_index('D####', inplace=True)
df_phase.reset_index(inplace=True)
df_phase.set_index('D####', inplace=True)
df_energy.reset_index(inplace=True)
df_energy.set_index('D####', inplace=True)
```

```python
df_traj_x
```

```python
df_traj_x.plot.bar(y='avg', yerr='std', ylabel='$\\langle X \\rangle$ [mm]',
                   fontsize=12, rot=90, figsize=(20, 6), alpha=0.8, color='b', xlabel='BPM')
```

```python
df_traj_y.plot.bar(y='avg', yerr='std', ylabel='$\\langle Y \\rangle$ [mm]',
                   fontsize=12, rot=90, figsize=(20, 6), alpha=0.8, color='r', xlabel='BPM')
```

```python
fig, ax = plt.subplots(figsize=(20, 6), dpi=180)
```

```python
df_traj_x.plot.bar(y='avg', yerr='std', ax=ax,
                   fontsize=14, rot=90, alpha=0.6, color='b', label='$\\langle x \\rangle$')
df_traj_y.plot.bar(y='avg', yerr='std', ax=ax,
                   fontsize=14, rot=90, alpha=0.6, color='r', label='$\\langle y \\rangle$')
ax.set_xlabel("BPM", fontsize=18, fontfamily='monospace')
ax.set_ylabel('Centroid Trajectory [mm]', fontsize=18, fontfamily='monospace')
[i.set_family('monospace') for i in ax.get_xticklabels()]
[i.set_family('monospace') for i in ax.get_yticklabels()]
```

```python
fig
```

```python
fig1, ax1 = plt.subplots(figsize=(20, 6), dpi=180)
```

```python
df_phase.plot.bar(y='avg', yerr='std', ax=ax1, label='$\\langle \phi \\rangle$',
                   fontsize=14, rot=90, alpha=0.6, color='g')
df_energy.plot.bar(y='avg', yerr='std', ax=ax1, label='$\\langle E \\rangle$', secondary_y=True,
                   fontsize=14, rot=90, alpha=0.6, color='m')
ax1.set_ylabel("Phase [deg]", fontsize=18, fontfamily='monospace')
ax1.right_ax.set_ylabel("Energy [MeV/u]", fontsize=18, fontfamily='monospace')
ax1.set_xlabel("BPM", fontsize=18, fontfamily='monospace')
[i.set_family('monospace') for i in ax1.get_xticklabels()]
[i.set_family('monospace') for i in ax1.get_yticklabels()]
```

```python
ax1.set_ylim(-200, 420)
ax1.right_ax.set_ylim(0, 60)
```

```python
fig1
```

# line plot

```python
fig3, ax3 = plt.subplots(figsize=(20, 6), dpi=180)
```

```python
v_avg = df_traj_x['avg']
v_std = df_traj_x['std']
ax3.fill_between(df_traj_x.index,
                 v_avg - 2 * v_std,
                 v_avg + 2 * v_std,
                 color='r', alpha=0.2)
df_traj_x.iloc[:,1:nshot+1].plot.line(ax=ax3, alpha=0.6, legend=False, lw=1)
df_traj_x.plot.line(y='avg', ax=ax3, alpha=0.9, color='b', lw=2, rot=90)
```
